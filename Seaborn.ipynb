```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

importing data

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv')
print(df)
```

Distribution Plot

```python
sns.set_context('talk' ,font_scale = 2.5)#styling
sns.set_style('white')#styling
sns.distplot(df['Attack'])
# if we use 'kde' then the line in graph will disappear and bins is used for width of bars
sns.distplot(df['Attack'] , kde = False , bins = 30)

sns.despine(left =False) # this is used to remove a paricular axis
```

Joint Plot

```python
sns.set_context('paper' ,font_scale = 1.5)
sns.set_style('dark')
sns.jointplot(x='Speed' , y='Attack' , data = df , kind = 'reg')

# diff kind og graph

sns.jointplot(x='Speed' , y='Attack' , data = df , kind = 'kde')

sns.jointplot(x='Speed' , y='Attack' , data = df , kind = 'hex')
```

KDE plot

```python
sns.set_context('paper' ,font_scale = 2.5)
sns.set_style('ticks')
plt.figure(figsize=(10,7))
sns.kdeplot(df['Speed'])
```

Pair Plot


```python
sns.set_style('whitegrid')
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
sns.pairplot(gas)
```

Pair plot in string

```python
sns.pairplot(df , hue = 'Legendary' , palette = 'Blues')
```

Rug Plot

```python
sns.rugplot(gas['USA'])
```

BAR Graph

```python
sns.barplot(x= 'Legendary' , y = 'Attack' , data = df , estimator = np.median)
sns.set_context('paper' , font_scale = 2.5)
```

count Plot

```python
sns.countplot(x = 'Legendary' , data = df)
```

Box Plot

```python
sns.boxplot(x = 'Type 1' , y = 'HP' , data = df , hue = 'Legendary') # 1 variable can be numeric and other can be string or both can be numeric
```

Voilin Plot

```python
plt.figure(figsize = (20,10))
sns.violinplot(x = 'Type 1' , y = 'HP' , data = df , hue = 'Legendary')
```

Strip plot

```python
plt.figure(figsize = (30,10))
sns.stripplot(x = 'Type 1' , y = 'HP' , data = df , jitter = True , hue = 'Legendary' , dodge = True)
# in above line jitter is used to separate the points , hue is added for male and female before that is was for both men and women , now dodge is used to separate male and female
```

swarm plot

```python
sns.swarmplot(x = 'Legendary' , y = 'HP' , data = df  , palette= 'magma')
plt.legend(loc=0)
```

Heat Map

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
plt.figure(figsize = (20 , 10))
gas_mx = gas.corr()
sns.heatmap(gas_mx , annot = True , cmap  = 'Blues' , linewidth = 1 , linecolor = 'red')
```

Cluster Map

```python
#Note : to implement cluster map all the string type coloumn should be deleted
# Legendary = df.pop('Legendary) , this is syntax to delete a particular column
Legendary = df.pop('Legendary')
Name = df.pop('Name')
Type_1 = df.pop('Type 1')
Type_2 = df.pop('Type 2')
sns.clustermap(df)
```

Pair Grid

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv')

df_g = sns.PairGrid(df , hue = "HP")
df_g.map(plt.scatter)
```

Facet Grid

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips_fg = sns.FacetGrid(tips , col = 'time' , hue = 'smoker')
tips_fg.map(plt.scatter , 'total_bill' , 'tip')
```

Regression plot

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , hue = 'sex' , data = tips , markers = ['o' , '^'] , scatter_kws ={'s' : 10 , 'linewidth' : 5.5})
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip', col = 'sex' , row = 'time' , data = tips)
```
